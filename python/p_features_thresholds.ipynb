```python
%pylab inline
import pyfits
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
mpl.rcParams['text.usetex']=True
```

```python
data=pyfits.open('ferengi_t01_data')
data=data[1].data
```

```python
target_zs=[0,.013,.02,.02475,.027]
def sb_list(list,z_lo,z_hi):
    return [row['mu_max_i'] for row in list if row['redshift'] > z_lo and row['redshift'] < z_hi]
def sim_z_list(list,z_lo,z_hi):
    return [row['sim_redshift'] for row in list if row['redshift'] > z_lo and row['redshift'] < z_hi]
```

```python
xedges=[.25,.35,.45,.55,.65,.75,.85,.95,1.05]
yedges=np.linspace(np.min(data['mu_max_i']),np.max(data['mu_max_i']),10)
extent=[xedges[0],xedges[-1],yedges[0],yedges[-1]]
```

```python
def hist(z_lo,target_z):
 return np.histogram2d(sim_z_list(data,z_lo,target_z),sb_list(data,z_lo,target_z),bins=(xedges,yedges))
```

```python
f=plt.figure(figsize=(13,6))
gs=gridspec.GridSpec(2,4)

vmin=0
vmax=250

Target_03=plt.subplot(gs[0,0])
plt.imshow(hist(target_zs[0],target_zs[1])[0].T,origin='lower',interpolation='nearest',extent=extent,aspect='auto',vmin=vmin,vmax=vmax)
plt.tick_params(labelbottom='off')
plt.ylabel('$\mathrm{Surface~Brightness~(mag/arcsec^2)}$',fontsize=12)
plt.title('$\mathrm{Target~z=0.3}$',fontsize=12)

Target_05=plt.subplot(gs[0,1],sharey=Target_03)
plt.imshow(hist(target_zs[1],target_zs[2])[0].T,origin='lower',interpolation='nearest',extent=extent,aspect='auto',vmin=vmin,vmax=vmax)
plt.tick_params(labelleft='off')
plt.tick_params(labelbottom='off')
plt.title('$\mathrm{Target~z=0.5}$',fontsize=12)


Target_08=plt.subplot(gs[1,0])
plt.imshow(hist(target_zs[2],target_zs[3])[0].T,origin='lower',interpolation='nearest',extent=extent,aspect='auto',vmin=vmin,vmax=vmax)
plt.xlabel('$\mathrm{Simulated~Redshift}$',fontsize=12)
plt.ylabel('$\mathrm{Surface~Brightness~(mag/arcsec^2)}$',fontsize=12)
plt.title('$\mathrm{Target~z=0.8}$',fontsize=12)


Target_1=plt.subplot(gs[1,1],sharey=Target_08)
plt.imshow(hist(target_zs[3],target_zs[4])[0].T,origin='lower',interpolation='nearest',extent=extent,aspect='auto',vmin=vmin,vmax=vmax)
plt.tick_params(labelleft='off')
plt.xlabel('$\mathrm{Simulated~Redshift}$',fontsize=12)
plt.title('$\mathrm{Target~z=1.0}$',fontsize=12)

All=plt.subplot(gs[:,2:])
plt.imshow(hist(target_zs[0],target_zs[4])[0].T,origin='lower',interpolation='nearest',extent=extent,aspect='auto',vmin=vmin,vmax=vmax)
plt.xlabel('$\mathrm{Simulated~Redshift}$',fontsize=12)
plt.title('$\mathrm{All~FERENGI}$',fontsize=12)

plt.colorbar()


plt.savefig('images/SB_vs_z.pdf')
```

```python
galaxies = set(data['sdss_id'])
```

```python
evos=[0,.5,1,1.5,2,2.5,3]
reds=[.3,.4,.5,.6,.7,.8,.9,1]
scatter_dct={}
for z in reds:
    for edge in yedges:
        scatter_dct[z,edge,'hi']=[]
        scatter_dct[z,edge,'lo']=[]
```

```python
for y in range(0,len(yedges)-1):
    for j,hi_z in enumerate(reds):    
        for i,g in enumerate(galaxies):
            this_gal=(data['sdss_id']==g)
            for row in data[this_gal]:
                if row['redshift']<.013:
                    for e in evos:
                        if row['sim_evolution']==e and round(row['sim_redshift'],2)==hi_z and row['mu_max_i'] > yedges[y] and row['mu_max_i'] < yedges[y+1]:
                            scatter_dct[hi_z,yedges[y],'hi'].append(row['p_features'])
                            for row in data[this_gal]:
                                if row['sim_evolution']==e and round(row['sim_redshift'],2)==.3:
                                    scatter_dct[hi_z,yedges[y],'lo'].append(row['p_features'])
         
```
