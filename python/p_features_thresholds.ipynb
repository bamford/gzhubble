```python
%pylab inline
import pyfits
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from operator import itemgetter
import scipy.stats

mpl.rcParams['text.usetex']=True
```

```python
#Load Data!
data=pyfits.open('../data/ferengi_debiasable_data')
data=data[1].data
```

```python
#cheap way of selecting only "Target z of 0.3" galaxies - ie those which were FERENGI'd from 0.3 to 1.0. 
#Target galaxies depend on redshift - Target 0.3 lie within z=0 and 0.013, Target 0.5 are between 0.013 and 0.02, etc. 
target_zs=[0,.013,.02,.0246,.027]
#Get list of SBs, simulated redshifts for galaxies within target redshift range
def sb_list(list,z_lo,z_hi):
    return [row['mu_max_i'] for row in list if row['redshift'] > z_lo and row['redshift'] < z_hi]
def sim_z_list(list,z_lo,z_hi):
    return [row['sim_redshift'] for row in list if row['redshift'] > z_lo and row['redshift'] < z_hi]
```

```python
#Defining bins to group data in redshift and surface brightness
xedges=[.25,.35,.45,.55,.65,.75,.85,.95,1.05]
yedges=np.linspace(np.min(data['mu_max_i']),np.max(data['mu_max_i']),10)
extent=[xedges[0],xedges[-1],yedges[0],yedges[-1]]
```

```python
def hist(z_lo,target_z):
 return np.histogram2d(sim_z_list(data,z_lo,target_z),sb_list(data,z_lo,target_z),bins=(xedges,yedges))
```

```python
#Histogram to check out space - find where galaxies are distributed in SB/redshift space and in what numbers

f=plt.figure(figsize=(13,6))
gs=gridspec.GridSpec(2,4)

vmin=0
vmax=250

#Galaxies redshifted from 0.3 - 1.0 
Target_03=plt.subplot(gs[0,0])
plt.imshow(hist(target_zs[0],target_zs[1])[0].T,origin='lower',interpolation='nearest',extent=extent,aspect='auto',vmin=vmin,vmax=vmax)
plt.tick_params(labelbottom='off')
plt.ylabel('$\mathrm{Surface~Brightness~(mag/arcsec^2)}$',fontsize=12)
plt.title('$\mathrm{Target~z=0.3}$',fontsize=12)

#Galaxies redshifted from 0.5 - 1.0
Target_05=plt.subplot(gs[0,1],sharey=Target_03)
plt.imshow(hist(target_zs[1],target_zs[2])[0].T,origin='lower',interpolation='nearest',extent=extent,aspect='auto',vmin=vmin,vmax=vmax)
plt.tick_params(labelleft='off')
plt.tick_params(labelbottom='off')
plt.title('$\mathrm{Target~z=0.5}$',fontsize=12)

#Galaxies redshifted from 0.8 - 1.0
Target_08=plt.subplot(gs[1,0])
plt.imshow(hist(target_zs[2],target_zs[3])[0].T,origin='lower',interpolation='nearest',extent=extent,aspect='auto',vmin=vmin,vmax=vmax)
plt.xlabel('$\mathrm{Simulated~Redshift}$',fontsize=12)
plt.ylabel('$\mathrm{Surface~Brightness~(mag/arcsec^2)}$',fontsize=12)
plt.title('$\mathrm{Target~z=0.8}$',fontsize=12)

#Galaxies only redshifted to 1.0
Target_1=plt.subplot(gs[1,1],sharey=Target_08)
plt.imshow(hist(target_zs[3],target_zs[4])[0].T,origin='lower',interpolation='nearest',extent=extent,aspect='auto',vmin=vmin,vmax=vmax)
plt.tick_params(labelleft='off')
plt.xlabel('$\mathrm{Simulated~Redshift}$',fontsize=12)
plt.title('$\mathrm{Target~z=1.0}$',fontsize=12)

#All galaxies
All=plt.subplot(gs[:,2:])
plt.imshow(hist(target_zs[0],target_zs[4])[0].T,origin='lower',interpolation='nearest',extent=extent,aspect='auto',vmin=vmin,vmax=vmax)
plt.xlabel('$\mathrm{Simulated~Redshift}$',fontsize=12)
plt.title('$\mathrm{All~FERENGI}$',fontsize=12)

plt.colorbar()
#plt.savefig('images/SB_vs_z.pdf')
```

```python
#Start with Target z=0.3 galaxies; these have p_features values spanning 0.3 to 1.0 
```

```python
#Pick out unique galaxies
galaxies = set(data['sdss_id'])
```

```python
#listing evolutions and redshifts 
evos=[0,.5,1,1.5,2,2.5,3]
reds=[.3,.4,.5,.6,.7,.8,.9,1]
#Defining lists of p_features at high and low (z=0.3) redshifts at given SB and redshifts. 
scatter_dct={}
for z in reds:
    for edge in yedges:
        scatter_dct[z,edge,'hi']=[]
        scatter_dct[z,edge,'lo']=[]
```

```python
#In each SB range and simulated redshift, make a list of p_features at that redshift and p_features at z=0.3
for y in range(0,len(yedges)-1):
    for j,hi_z in enumerate(reds):    
        for i,g in enumerate(galaxies):
            this_gal=(data['sdss_id']==g)
            for row in data[this_gal]:
                if row['redshift']<.013:
                    for e in evos:
                        if row['sim_evolution']==e and round(row['sim_redshift'],2)==hi_z and row['mu_max_i'] > yedges[y] and row['mu_max_i'] < yedges[y+1]:
                            scatter_dct[hi_z,yedges[y],'hi'].append(row['p_features'])
                            for row in data[this_gal]:
                                if row['sim_evolution']==e and round(row['sim_redshift'],2)==.3:
                                    scatter_dct[hi_z,yedges[y],'lo'].append(row['p_features'])
```

```python
#Plot p_features at a given redshift vs p_features at z=0.3 at each redshift, binned by surface brightness
f=figure(figsize=(10,8))
totalcount=0
badcount=0
gs=gridspec.GridSpec(9,10)
gs.update(wspace=0)
gs.update(hspace=0)
yedge_int=[8,7,6,5,4,3,2,1,0]
p_features_bins=np.linspace(.05,.95,10)
y_label=[]
for i in range(0,len(yedges)-1):
    y_label.append(round((yedges[i]+yedges[i+1])/2,1))
y_label=y_label[::-1]
for j,y in enumerate(yedge_int):
    for i,z in enumerate(reds):
        plt.subplot(gs[j,i])
        xs=scatter_dct[z,yedges[y],'lo']
        ys=scatter_dct[z,yedges[y],'hi']
        totalcount+=len(xs) 
   #     xs,ys=[list(x) for x in zip(*sorted(zip(xs, ys), key=itemgetter(0)))] #sorting only relevant so the fits can be plot later 
        this_dct={}
        for k in range(0,10):
            this_dct[k/10.,(k+1)/10.]=[]
        for k in range (0,10):
            for l,val in enumerate(ys):
                if ys[l] >= k/10. and ys[l] < (k+1)/10.:
                    this_dct[k/10.,(k+1)/10.].append(xs[l]) 
        plt.scatter(xs,ys,marker='.',s=.5,alpha=.2)
        for k in range(0,10):
            plt.axhline(y=k/10.,alpha=.1,lw=.5)
        p_features_means=[]
        x_error_lo=[]
        x_error_hi=[]
        for k in range(0,10):
            p_features_means.append(np.mean(this_dct[k/10.,(k+1)/10.]))
            try:
                x_error_lo.append(p_features_means[k] - np.percentile(this_dct[k/10.,(k+1)/10.],25))
            except ValueError:
                x_error_lo.append(0)
            try:
                x_error_hi.append(np.percentile(this_dct[k/10.,(k+1)/10.],75) - p_features_means[k])
            except ValueError:
                x_error_hi.append(0)
        plt.errorbar(p_features_means,p_features_bins,xerr=[x_error_lo,x_error_hi],fmt='o',c='r',markersize=1)
        for k in range(0,10):
            try:
                if x_error_lo[k]+x_error_hi[k] > 0.2 and len(this_dct[k/10.,(k+1)/10.])>5: # if width of inter quartiles > .2 and if>5 pts, shade that shit 
                    plt.axhspan(k/10.,(k+1)/10.,alpha=.05)
                    badcount+=len(this_dct[k/10.,(k+1)/10.])
            except ValueError:
                pass
        #janky plotting the middle spaces green, should probably automate at some point. yay brute force on a train. 
        if z ==.6 and j==2:
            plt.axhspan(.3,.4,alpha=.05,facecolor='g')
        if z ==.6 and j==4:
            plt.axhspan(.2,.3,alpha=.05,facecolor='g')
        if z ==.7 and j==3:
            plt.axhspan(.3,.4,alpha=.05,facecolor='g')
        if z ==.7 and j==4:
            plt.axhspan(.2,.3,alpha=.05,facecolor='g')
        plt.tick_params(labelleft='off')
        plt.tick_params(top='off',bottom='off',right='off',left='off',labelbottom='off')
        plt.xlim(0,1)
        plt.ylim(0,1)
        if j==8:
            plt.xlabel('%s'%reds[i],fontsize=12)
        if i==0:
            plt.ylabel('%s'%y_label[j],fontsize=12)
plt.subplot(gs[4,9])
plt.xlabel(r'$p_{features},z=0.3$')
plt.ylabel(r'$p_{features}$')
f.text(.05,.5,r'$\mathrm{\mu_{max,i}~(mag/arcsec^2)}$',rotation=90,fontsize=20,va='center')
f.text(.45,.05,r'$\mathrm{redshift}$',fontsize=20,ha='center')

#plt.savefig('images/p_vs_p_SB_redshift_2.pdf')
```
